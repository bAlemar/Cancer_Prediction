```python
#Checking frequency of variables
#This is better for categorical data, so ignore this one
def freq_data(df):
    df_2 = df.copy()
    dict_len = {'feature':[],'dtypes':[],'len(unique)':[],
                'most_frequent':[],'most_frequent(name)':[]}
    for i in df_2.columns:
        dict_len['feature'].append(i)
        dict_len['dtypes'].append(df[i].dtypes)
        dict_len['len(unique)'].append(len(df_2[i].unique()))
        dict_len['most_frequent(name)'].append(df_2[i].value_counts(100).sort_values(ascending=False).index[0])
        dict_len['most_frequent'].append(df_2[i].value_counts(100).sort_values(ascending=False).values[0])
        df_feature = pd.DataFrame(dict_len)
        df_feature = df_feature.sort_values(by='len(unique)',ascending=False)
    return df_feature
df_feature = freq_data(df)
print(df_feature.drop(columns=['dtypes']).sort_values('most_frequent',ascending=False))
```

```python
#Test_train_split
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder,StandardScaler
X = df.drop(columns=['diagnosis','id'])
y = df['diagnosis']
le = LabelEncoder()
y = le.fit_transform(y)
columns_name = X.columns
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state= 0,test_size=0.3)
#Não muda nada,Desicion Tree,geralmente não precisa 
#sc = StandardScaler()
#X_train = sc.fit_transform(X_train)
#X_test = pd.DataFrame(sc.transform(X_test),columns=columns_name)
```

```python
## Treinando o Modelo
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(random_state = 0,max_depth=5)
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
```

```python
## Avaliando modelo,p
from sklearn.metrics import classification_report,accuracy_score,mean_absolute_error,roc_auc_score,roc_curve,plot_roc_curve
report = classification_report(y_test,y_pred) 
print(report)
print('Mean Absolute Error: ', mean_absolute_error(y_pred, y_test))
print('acurácia treino',accuracy_score(y_train,model.predict(X_train)))
print('acurácia test',accuracy_score(y_test,y_pred))
fp_rate, tp_rate, thresholds = roc_curve(y_test, y_pred)
print('AUC',roc_auc_score(y_test,y_pred))
plot_roc_curve(model,X_test,y_test)
```

```python
def Analise_Feature(model,X_test):
    aux_dic = {'name':[],'coef':[]}
    pd.set_option('display.max_rows',len(X_test.columns))
    for name, importance in zip(X_test.columns,model.feature_importances_):
        aux_dic['name'].append(name)
        aux_dic['coef'].append(importance)
        df_features_importance = pd.DataFrame(aux_dic)
    print(df_features_importance.sort_values(by='coef',ascending=False))
    return df_features_importance
Analise_Feature(model,X_test)
```

```python
## Aplying HiperFeatures and HiperParameters
## Tentando Aumentar AUC
```

```python
from sklearn.decomposition import PCA
max_features = len(X.columns)
dict_results = {'n_components':[],'AUC':[]}
#pca = PCA(n_components=i)
for i in range(1,max_features):
    pca = PCA(n_components=i)
    X_train,X_test,y_train,y_test = train_test_split(X,y,random_state= 0,test_size=0.3)
    X_train = pca.fit_transform(X_train)
    X_test = pca.transform(X_test)
    model = DecisionTreeClassifier(random_state=0)
    model.fit(X_train,y_train)
    y_pred = model.predict(X_test)
    AUC = roc_auc_score(y_test,y_pred)
    dict_results['n_components'].append(i)
    dict_results['AUC'].append(AUC)
plt.plot(dict_results['n_components'],dict_results['AUC'])
plt.show()
dict_pca_results = pd.DataFrame(dict_results)
print(dict_pca_results.sort_values('AUC',ascending=False).head(-1))
```

```python
#N_Components = 11
#Agora utilizaremos GridSearch para Hiperparametros:
from sklearn.model_selection import GridSearchCV
pca = PCA(n_components=12)
param_grid = { 'max_depth': [1,2,3,5,10,15,20,None],
             'min_samples_split':[1,2,3,5,10,15,20,None],
             'min_samples_leaf':[1,2,3,5,10,15,20,None],
             'random_state':[0]}

X_train,X_test,y_train,y_test = train_test_split(X,y,random_state= 0,test_size=0.3)
X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)
model = DecisionTreeClassifier(random_state=0)
model = GridSearchCV(model,param_grid,scoring='roc_auc',n_jobs=-1,cv=5)
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
y_pred_2 = model.predict(X_train)

print('Param for GS',model.best_params_)
print('CV score for GS',model.best_score_)
print('Train AUC ROC:',roc_auc_score(y_train,y_pred_2))
print('Test AUC ROC:',roc_auc_score(y_test,y_pred))
```

```python
#Basicamente copiamos e colamos o processo que fiz para escolha do PCA, só que agora acrescentamos
#o GridSearch
from sklearn.decomposition import PCA
max_features = len(X.columns)
dict_results_PCA = {'n_components':[],'AUC':[]}
param_grid = { 'max_depth': [1,2,3,5,10,15,20,None],
             'min_samples_split':[1,2,3,5,10,15,20,None],
             'min_samples_leaf':[1,2,3,5,10,15,20,None],
             'random_state':[0]}
#pca = PCA(n_components=i)
for i in range(1,max_features):
    pca = PCA(n_components=i)
    X_train,X_test,y_train,y_test = train_test_split(X,y,random_state= 0,test_size=0.3)
    X_train = pca.fit_transform(X_train)
    X_test = pca.transform(X_test)
    model = DecisionTreeClassifier(random_state=0)
    model = GridSearchCV(model,param_grid,scoring='roc_auc',n_jobs=-1,cv=5)
    model.fit(X_train,y_train)
    y_pred = model.predict(X_test)
    AUC = roc_auc_score(y_test,y_pred)
    dict_results_PCA['n_components'].append(i)
    dict_results_PCA['AUC'].append(AUC)
plt.plot(dict_results_PCA['n_components'],dict_results_PCA['AUC'])
plt.show()
dict_pca_gridSearch = pd.DataFrame(dict_results_PCA)
print(dict_pca_gridSearch.sort_values('AUC',ascending=False).head(-1))
```
